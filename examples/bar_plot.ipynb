## Bar Plots with MatPlus

### Simple Bar Plot

```python
import numpy as np
from MatPlus import BarPlot

# Create data
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 7, 12, 9])

# Create and display basic bar plot
bar = BarPlot(x, y)
bar.plot()
```

### Advanced Line Plots

```python
import numpy as np
from MatPlus import LinePlot

# Create multiple line data
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

# Multiple line plot
line = LinePlot([x1, x2], [y1, y2])
line.plot()

# Line plot with custom axis limits and line width
x = np.linspace(-5, 5, 100)
y = x**2
line = LinePlot(x, y, lowerlimx=-6, upperlimx=6, lowerlimy=-1, upperlimy=30, lw=2)
line.plot()
```
