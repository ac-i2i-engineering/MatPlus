## Bar Plots with MatPlus

### Simple Bar Plot

```python
import numpy as np
from MatPlus import BarPlot

# Create data
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 7, 12, 9])

# Create and display basic bar plot
bar = BarPlot(x, y)
bar.plot()
```

### Advanced Bar Plot

```python
import numpy as np
from MatPlus import BarPlot

# Create more complex data
categories = np.array([1, 2, 3, 4, 5])
values = np.array([23, 17, 35, 29, 12])

# Create a customized bar plot with:
# - Custom axis limits
# - Thinner bars (width=0.6)
# - Thicker outlines (linewidth=2)
advanced_bar = BarPlot(
    x=categories,
    y=values,
    lowerlimx=0.5,    # Start x-axis at 0.5
    lowerlimy=0,      # Start y-axis at 0
    upperlimx=5.5,    # End x-axis at 5.5
    upperlimy=40,     # End y-axis at 40
    wd=0.6,           # Make bars narrower than default
    lw=2,              # Make bar outlines thicker
    title= "advanced bar plot",
    xlabel="categories",
    ylabel="values"
)

# Plot the customized bar chart
advanced_bar.plot()
```
