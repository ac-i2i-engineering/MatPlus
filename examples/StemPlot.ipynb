### Simple Stem Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from MatPlus.StemPlot import StemPlot

x = np.arange(0, 10)
y = np.sin(x)

# Create and display a basic stem plot
basic_stem = StemPlot(x, y)
basic_stem.plot()
```

### Advanced Stem Plot

```python
# Create data
x = np.arange(-20, 30, 0.5)
y = 20 * np.cos(x)

# Create a customized stem plot
custom_stem = StemPlot(
    x, y,
    linefmt='--r',  # Red dashed lines
    markerfmt='*b',  # Blue star markers
    basefmt='-g',    # Green baseline
    label='Cosine'   # Add a label
)
custom_stem.plot()

# Create data
x = np.arange(1, 6)
y = x**2

# Create a horizontal stem plot
horizontal_stem = StemPlot(
    x, y,
    orientation='horizontal',
    linefmt='-g',
    markerfmt='Dg',  # Green diamond markers
    label='Squared values',
    lowerlimx=-3, # Manually set the x and y limits
    upperlimx=5,
    upperlimy=8,
    lowerlimy=-1,
)
horizontal_stem.plot()
```
