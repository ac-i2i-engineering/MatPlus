# Line Plots with MatPlus

### Simple Line Plot

```python
import numpy as np
from MatPlus import LinePlot

# Create some simple data
x = np.linspace(0, 10, 50)
y = x * 0.5 + 2  # Simple linear function
# Create and display a basic line plot
simple_line = LinePlot(x, y)
simple_line.plot()
```

### Advanced Line Plot

```python
import numpy as np
from MatPlus import LinePlot

# Create multiple line data
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

# Multiple line plot
line = LinePlot([x1, x2], [y1, y2])
line.plot()

# Line plot with custom axis limits and line width
x = np.linspace(-5, 5, 100)
y = x**2
line = LinePlot(x, y, lowerlimx=-6, upperlimx=6, lowerlimy=-1, upperlimy=30, lw=2, width = 2, height = 3)

line.plot()
```
