## Histograms with MatPlus

### Simple Histogram

```python
import numpy as np
from MatPlus import Histogram

# Create random data
data = np.random.normal(0, 1, 1000)

# Create and display a basic histogram
hist = Histogram(data)
hist.plot()
```

### Advanced Histogram

```python
import numpy as np
from MatPlus import Histogram

# Create data
data = np.concatenate([np.random.normal(-2, 1, 300), np.random.normal(2, 1, 300)])

# Create histogram with custom bins, density normalization and color
hist = Histogram(data, bins=20, density=True, color="blue")
hist.plot()

# Histogram with custom bin edges
custom_bins = [-5, -3, -1, 0, 1, 3, 5]
hist = Histogram(data, bins=custom_bins, color="green")
hist.plot()
```
